# Homework clase 4

A partir del CSV hospitales2.csv, generar un archivo CSV de salida, que contenga los siguientes campos en este orden:
- latitude
- longitude
- name
- label

Con los correspondientes datos extraídos del CSV original, el campo name tiene que corresponder con la dirección del hospital, y el campo label con el nombre del hospital.


Ejemplo de visualización en: https://www.gpsvisualizer.com

```python
import pandas as pd
import numpy as np
```

```python
hospitales = pd.read_csv('hospitales2.csv')
```

```python
hospitales.head()
```

```python
hospitales.info()
```

```python
hospitales.columns
```

```python
hospitales['WKT'][0]
```

```python
type(hospitales['WKT'][0])
```

```python
coordenadas = hospitales['WKT'][0].split()

print(coordenadas)
```

```python
coordenadas[1].strip('(')
```

```python
float(coordenadas[2].strip(')'))
```

```python
lista_coords = [coordenadas.split()  for coordenadas in hospitales['WKT']]
```

```python
print(lista_coords)
```

```python
longitud = [float(long[1].strip('(')) for long in lista_coords]
```

```python
latitud  = [float(long[2].strip(')')) for long in lista_coords]
```

```python
hospitales['latitude'] = latitud
```

```python
hospitales['longitude'] = longitud
```

```python
hospitales.head()
```

```python
hospitales.columns
```

```python
hospitales.rename(columns={'NOMBRE': 'label', 'DOM_NORMA': 'name'}, inplace=True)
```

```python
hospitales.head()
```

```python
columnas = ['latitude', 'longitude', 'name', 'label']
```

```python
hospitales_final = hospitales[columnas]
```

```python
hospitales_final
```

```python
hospitales_final.to_csv('hospitales_cleaned.csv', index=False)
```

```python
hospitales_final.to_excel('hospitales_final.xlsx')
```
