```python
dicc_emisiones = { 
    'cod_pais' : [],
    'nom_pais' : [],
    'region' : [],
    'anio' : [],
    'co2' : [],
    'co2_percapita' : []
}
```

```python
next(archivo)

for i in archivo:
    linea = i.strip()
    campos = linea.split('|')
    dicc_emisiones['cod_pais'].append(campos[0])
    dicc_emisiones['nom_pais'].append(campos[1])
    dicc_emisiones['region'].append(campos[2])
    dicc_emisiones['anio'].append(campos[3])
    dicc_emisiones['co2'].append(campos[4])
    dicc_emisiones['co2_percapita'].append(campos[5])
```

```python
region = 'América Latina y Caribe'
```

```python
emisiones = 0
```

```python
for idx, val in enumerate(dicc_emisiones['region']):
    if val == region and dicc_emisiones['anio'][idx] == anio and isinstance(dicc_emisiones['co2'][idx], float):
        emisiones += dicc_emisiones['co2'][idx]
```

```python
print('Total de emisiones de co2 para AL y Caribe en 2010:\n\n\t', emisiones)
```
