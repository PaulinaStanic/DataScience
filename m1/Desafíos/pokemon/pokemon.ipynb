```python
pokedex = pd.merge(a, status, left_on='ndex', right_on='podedex_index') # mergea el archivo a, con el archivo pokemon faltante, matcheando 'ndex' con 'podedex_index'
```

```python
pokedex= pokedex.sort_values(by='ndex') # ordenamos la información de manera creciente teniendo 'ndex' como parámetro
```

```python
pokedex.head(15)
```

```python
pokedex['ndex'] = pokedex['ndex'].astype(int) # convertimos 'ndex' en int
```

```python
pokedex.columns # para ver columnas duplicadas
```

```python
pokedex= pokedex.drop(columns=['podedex_index', 'name_y', 'name', 'Unnamed: 0']) # eliminamos columnas duplicadas
```

```python
pokedex.columns
```

```python
print(sum(pokedex['pokemon_weight_kg'].isnull()))
```

```python
print(sum(pokedex['pokemon_weight_grs'].isnull()))
```

```python
pokedex.shape
```

```python
pokedex['pokemon_weight']= pokedex['pokemon_weight_grs']/1000 # dividimos el peso en gramos en 1000 para tener el peso en kg
```

```python
pokedex['weight']= pokedex['pokemon_weight_kg'].where(pokedex['pokemon_weight'].isna(),pokedex['pokemon_weight']) # en donde el peso de pokemon_weight_kg no tenga valor, se tomará el valor de pokemon_weight
```

```python
pokedex= pokedex.drop(columns=['pokemon_weight_kg', 'pokemon_weight_grs', 'pokemon_weight']) # eliminamos columnas que utlizamos para procesar el peso
```

```python
pokedex.columns
```

```python
pokedex= pokedex.drop_duplicates(subset=['ndex']) # eliminamos los duplicados teniendo como parámetro la columna 'ndex'
```
